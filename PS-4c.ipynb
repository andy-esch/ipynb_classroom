## #2 -- A $43.8 kg$ sign is suspended by two wires, at angles of $43^{\circ}$ and $55.0^{\circ}$.  Find the tension in wires 1 and 2.

```python
import numpy as np
from IPython.core.display import Image
Image('http://www.apsva.us/cms/lib2/VA01000586/Centricity/domain/3724/intensified/ps-images/ps-4c-2.png')
```

```python
m = 43.8; g = 9.81
th1 = np.radians(43); th2 = np.radians(55);
```

Sum of forces:
$$\Sigma F_{x} = T_{2}\cos\theta_{2} - T_{1}\cos\theta_{1} = 0$$
$$\Sigma F_{y} = T_{1}\sin\theta_{1} + T_{2}\sin\theta_{2} - mg = 0$$
    
Therefore, by $\Sigma F_{x}$,
$$T_{1} = T_{2} \frac{\cos\theta_{2}}{\cos\theta_{1}}$$

$\Sigma F_{y}$ then becomes,
$$T_{2}\sin\theta_{2} + T_{2}\frac{\cos\theta_{2}}{\cos\theta_{1}}\sin\theta_{1} = mg$$
Simplifying, this becomes:
$$T_{2}\cos\theta_{2}\,\left(\tan\theta_{2} + \tan\theta_{1}\right) = mg$$

$$T_{2} = \frac{\sec\theta_{2}}{\tan\theta_1 + \tan\theta_2}mg$$
where $\sec\theta = \frac{1}{\cos\theta}$


Likewise, solving for $T_{1}$,
$$T_{1} = \frac{\sec\theta_{1}}{\tan\theta_1 + \tan\theta_2}mg$$

In code, T1 and T2 can be written as a single function, where the order of angles determines whether tension 1 or .

```python
# secant function used below
def sec(ang):
    return 1.0 / np.cos(ang)

def T(ang1, ang2):
    return (sec(ang1) / (np.tan(ang1) + np.tan(ang2))) * m * g
```

```python
T1 = T(th1,th2);
T2 = T(th2,th1);
print "T1 = %4.0f N, T2 = %4.0f N" % (T1, T2)
print "The object has a weight of %4.0f N" % (m*g)
```

### Sanity Checks

```python
print "Do forces balance in the x-direction?"
print "T1x = %4.0f N" % (-1.0 * T1 * np.cos(th1))
print "T2x = %4.0f N" % (T2 * np.cos(th2))
```

```python
print "Do the y-components of the tension forces equal the weight (430 N)?"
print "T1y + T2y = %4.0f + %4.0f = %4.0f N" % ((T1 * np.sin(th1)),
                                               (T2 * np.sin(th2)),
                                               (T1 * np.sin(th1) + T2 * np.sin(th2)))
```
